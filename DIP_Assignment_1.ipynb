### Task1

```python
#loading image
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/magnet.jpg")
cv2_imshow(img)
```

```python
#resizing to 256 by 256 pixels by cubic interpolation
resize_img = cv2.resize(img, (256,256), interpolation = cv2.INTER_CUBIC)
cv2_imshow(resize_img)
```

### Task2

```python
#converting RGB to grayscale
gray_img = cv2.cvtColor(resize_img, cv2.COLOR_RGB2GRAY)
cv2_imshow(gray_img)
```

### Task3

```python
#converting grayscale to binary
binary_img = cv2.threshold(gray_img, 100, 255, cv2.THRESH_BINARY)
cv2_imshow(binary_img[1])
```

### Task4

```python
#loading image
coin_img = cv2.imread("/content/coins.png")
cv2_imshow(coin_img)
```

```python
#removing noise and smoothing the details which will help in segment the coins as a single white circle
coin_img_enhanced = cv2.fastNlMeansDenoisingColored(coin_img, None, 50, 50, 50, 7)
cv2_imshow(coin_img_enhanced)
```

```python
#converting to grayscale and then to binary
gray_img = cv2.cvtColor(coin_img_enhanced, cv2.COLOR_RGB2GRAY)
binary_img = cv2.threshold(gray_img, 130, 255, cv2.THRESH_BINARY_INV)
cv2_imshow(binary_img[1])
```

```python
#counting the connected components
#subtracting 1 from ret to exclude the label '0' of background
ret, labels = cv2.connectedComponents(binary_img[1])
print("Total coins:", ret-1)
```
